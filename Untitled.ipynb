```python
import nltk
import nltk.sentiment as se
import xml.etree.ElementTree as ET
f=open(r'ABSA16_Laptops_Train_SB1_v2.xml','r')
data=f.read()
dt=ET.parse('ABSA16_Laptops_Train_SB1_v2.xml')
root=dt.getroot()
ip=[]
op=[]
for child in root:
    for child1 in child:
        for child2 in child1:
            i=0
            t=[]
            for child3 in child2:
                if i==0:
                    #print(child3.text)
                    ip.append(child3.text)
                else:
                    for child4 in child3:
                        #print(child4.attrib)
                        t.append(child4.attrib)
                i=i+1
            op.append(tuple(t))
            
```

```python
from nltk.corpus import stopwords
import nltk
# tokenizing
token_ip=[nltk.word_tokenize(s) for s in ip]    
# removing stopwords
token_ip=[[w for w in s if w not in stopwords.words('english')] for s in token_ip]
# pos tagging
pos_tagged=[nltk.pos_tag(s) for s in token_ip]
```

```python
def collect_nouns(text):
    l=len(text)
    lst=[]
    for i in range(l):
        for j in range(len(text[i])):
            if text[i][j][1]=='NN':
                lst.append(text[i][j][0])
        
    return lst    
```

```python
nouns=collect_nouns(pos_tagged)
numpy.shape(nouns)
```

```python
def collect_verb(text):
    l=len(text)
    lst=[]
    for i in range(l):
        for j in range(len(text[i])):
            if text[i][j][1]=='VB':
                lst.append(text[i][j][0])
        
    return lst 
```

```python
verbs=collect_verb(pos_tagged)
numpy.shape(verbs)
```

```python
def collect_adverb(text):
    l=len(text)
    lst=[]
    for i in range(l):
        for j in range(len(text[i])):
            if text[i][j][1]=='RB':
                lst.append(text[i][j][0])
        
    return lst 
```

```python
def collect_adjectives(text):
    l=len(text)
    lst=[]
    for i in range(l):
        for j in range(len(text[i])):
            if text[i][j][1]=='JJ':
                lst.append(text[i][j][0])
        
    return lst 
```

```python
def collect_two_noun(text):
    l=len(text)
    lst=[]
    for i in range(l):
        for j in range(len(text[i])-1):
            if text[i][j][1]=='NN' and text[i][j+1][1]=='NN':
                tmp=str()
                tmp=tmp+str(text[i][j][0])+' '
                tmp=tmp+str(text[i][j+1][0])
                lst.append(tmp)
        
    return lst 
```

```python
two_nouns=collect_two_noun(pos_tagged)
```

```python
def get_frequency(two_nouns):
    freq=[]
    for w in two_nouns:
        tmp=[]
        tmp.append(w)
        tmp.append(two_nouns.count(w))
        freq.append(tmp)
    return freq
```

```python
p=get_frequency(two_nouns)
```

```python
freq_2=[]
for w in p:
    if w[1]>=5:
        freq_2.append(w)
```
